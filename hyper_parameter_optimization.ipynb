```python
result = hpo_pipeline(
    model = "TransE",  #name of best model،
    optimizer = "Adam",#name of best optimizer
    loss = "MSE",#best Loss
    dataset = "nations",  # custom dataset
    training_loop = "sLCWA",
    n_trials = 10,#number of trials


    model_kwargs_ranges=dict(embedding_dim=dict(           
            type='categorical',
            choices = [100, 200],
        )),
    pruner="MedianPruner",
    sampler ="tpe",
    training_kwargs_ranges=dict(num_epochs =dict(  # range of epoch and batchsize
                type=int,
                low = 10,
                high = 20,
                q = 5,),
                batch_size=dict(
                type='categorical',
                choices= [64,128],             
                    ),),
    negative_sampler_kwargs_ranges=dict(
            num_negs_per_pos=dict(
                type=int,
                low = 1,
                high = 11,
                q = 10,
            ),),
    
    optimizer_kwargs_ranges=dict(lr=dict(
                type = float,
                low = 0.01,
                high =0.1,),),
    dataset_kwargs = {"create_inverse_triples":False},
    model_kwargs = {"random_seed":42},
    evaluator_kwargs={"filtered": True},
    evaluation_kwargs={ 'batch_size':1},
    # storage='sqlite:///../data/kg/sql/hpo_results.db',  # <-- This line enables resuming
    # study_name='hpo_study',               # <-- This names the study
    # load_if_exists=True                   # <-- Resume if interrupted
)
#print result of hpo
print(result)

#save result file
result.save_to_directory('./nationarg1')
```
