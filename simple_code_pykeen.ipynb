```python
#import pykeen library, and pipline from pykeen, and json for type of output file
import pykeen
from pykeen.pipeline import pipeline
import json
```

```python
#we fix path for save result
path = "./res"
```

```python
#in this code we create a knowledge graph model for nations dataset
result_of_model = pipeline(
    dataset = "nations", #we can use another dataset
    model = "TransE", #we can use another model 
    epochs = 10
)
```

```python
#we can save result 
result_of_model.save_to_directory(path)
```

```python
#for see result of metrics and loss, we can run this code
with open(path + '/results.json', 'r') as file:
    data = json.load(file)
data
```
